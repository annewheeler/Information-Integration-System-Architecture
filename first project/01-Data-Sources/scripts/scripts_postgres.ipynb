```python
import pandas as pd

df_pg = pd.read_csv('01_HR_Identity_PG.csv')

# Extragem departamentele unice și le dăm un ID
depts = pd.DataFrame(df_pg['Department'].unique(), columns=['dept_name'])
depts.index.name = 'dept_id'
depts.to_csv('pg_departments.csv')

# Extragem domeniile de educație unice
edu = pd.DataFrame(df_pg['EducationField'].unique(), columns=['field_name'])
edu.index.name = 'edu_id'
edu.to_csv('pg_education_fields.csv')

# Pregătim tabelul de angajați înlocuind numele cu ID-uri (Mapping)
# Asta demonstrează conceptul de Foreign Key
dept_map = {name: i for i, name in enumerate(df_pg['Department'].unique())}
edu_map = {name: i for i, name in enumerate(df_pg['EducationField'].unique())}

df_pg['dept_id'] = df_pg['Department'].map(dept_map)
df_pg['edu_id'] = df_pg['EducationField'].map(edu_map)

# Păstrăm doar coloanele necesare pentru tabelul final
employees_final = df_pg[['EmployeeNumber', 'Age', 'Gender', 'MaritalStatus', 'dept_id', 'edu_id']]
employees_final.to_csv('pg_employees_final.csv', index=False)
```
